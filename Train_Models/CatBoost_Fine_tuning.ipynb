```python
df_kaggle = pd.read_csv('../datasets/kaggle_data_transfromed.csv')
df_kaggle.head()
```

```python
df_empty = pd.DataFrame(columns=df_kaggle.columns)
```

```python
columns_empty = df_empty.columns
for column in columns_empty:
    if column in df_survey.columns:
        df_empty[column] = df_survey[column]
    
df_empty.head()
```

```python
# Define the logic to fill the 'Study Degree_Unknown' column
def fill_study_degree_unknown(row):
    if row['Study Degree_High School'] == 0 and row['Study Degree_Master’s Degree'] == 0 and row['Study Degree_PhD'] == 0:
        return 1
    else:
        return 0

# Apply the function to each row in the df_empty dataframe
df_empty['Study Degree_Unknown'] = df_empty.apply(fill_study_degree_unknown, axis=1)
df_empty.head()
```

```python
import numpy as np
import random


profession_columns = [col for col in df_empty.columns if col.startswith('Profession')]
# Iterate over each row in the dataframe
for index, row in df_empty.iterrows():
    # Check if 'Profession_Student' or 'Profession_Unknown' are 1
    if row['Profession_Student'] == 1 or row['Profession_Unknown'] == 1:
        # If either is 1, set all other profession columns to 0
        for col in profession_columns:
            df_empty.at[index, col] = 0
    else:
        # If both are 0, randomly set one profession column (other than 'Profession_Student' and 'Profession_Unknown') to 1
        random_profession = random.choice(profession_columns)
        for col in profession_columns:
            if col == random_profession:
                df_empty.at[index, col] = 1
            else:
                df_empty.at[index, col] = 0
df_empty.head()
```
