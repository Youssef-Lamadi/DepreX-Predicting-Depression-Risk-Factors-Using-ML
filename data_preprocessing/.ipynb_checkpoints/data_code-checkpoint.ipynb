```python
print("----------Numerical Featuers---------")
num_feats = df.select_dtypes(include="float64").columns
num_feats
```

```python
# Import necessary library
import pandas as pd

# Assuming df is your DataFrame
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Feature: {column}")
    print(f"Unique Values ({len(unique_values)}): {unique_values}\n")
```

```python
fig = plt.figure(figsize=(10, 18))
for i, feature in enumerate(obj_feats):
    ax = fig.add_subplot(5, 2, i + 1)
    
    df_temp = df[feature].value_counts().nlargest(20).reset_index()
    df_temp.columns = [feature, 'count']  # Rename columns for clarity

    sns.barplot(data=df_temp, y=feature, x='count', color="#1E90FF", orient="h", alpha=0.9, ax=ax)

    ax.bar_label(
        ax.containers[0], label_type="center", padding=5, color="black", fontsize=8
    )

    ax.set_ylabel(feature, fontsize=12)
    ax.set_xlabel("Counts", fontsize=12)
    ax.tick_params(axis="both", labelsize=10)

fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
for i, feature in enumerate(num_feats):
    ax = fig.add_subplot(7, 2, i + 1)
    sns.histplot(
        data=df,
        x=feature,
        stat="percentage",
        kde=True,
        bins=20,
        alpha=0.4,
        color="#1E90FF",
        line_kws={"lw": 2, "linestyle": "solid", "color": "#010a59"},
        ax=ax,
    )
    plt.tick_params(axis="both", labelsize=9)
    ax.set_ylabel("Percent", fontsize=9)
    ax.set_xlabel(feature, fontsize=9)
fig.tight_layout()
plt.show()
```

### Target Variable Distribution

```python
target_colors = [
    "#e32020",
    "#6fcb9f",
]
df["Depression"].value_counts().plot.pie(
    figsize=(6, 6),
    autopct="%.0f%%",
    colors=target_colors,
    explode=[0.03] * len(df["Depression"].value_counts()),
    textprops={"fontsize": 15, "color": "#000000"},
    title="Target Variable Distribution",
)
plt.ylabel("")  # Removes y-axis label for a cleaner pie chart
plt.show()
```

#### `The 89%/11% class distribution is an important insight that suggests a significant class imbalance. This should be taken into consideration during preprocessing and model training to improve the model's ability to correctly predict the minority class. Techniques like resampling, changing evaluation metrics, and adjusting algorithm parameters can help make the model more effective at handling imbalanced data.`

```python
# Visualize missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# 3. Categorical Features
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("\nCategorical Columns:", categorical_columns)

for col in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col, palette="viridis")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 4. Numerical Features
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumerical Columns:", numerical_columns)

for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, kde=True, color="blue")
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# 5. Correlations (Numerical Data)
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# 6. Target Variable Analysis
# Customize this part for your specific target variable
target = "target_column_name"  # Replace with your target column name
if target in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=target, palette="viridis")
    plt.title(f"Distribution of Target Variable: {target}")
    plt.show()
```

```python
# 7. Outlier Detection (Boxplots)
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x=col, color="red")
    plt.title(f"Outliers in {col}")
    plt.show()
    
```
