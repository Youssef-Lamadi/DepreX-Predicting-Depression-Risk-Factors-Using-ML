### 1. Univariate Analysis

```python
depression_counts = df['Depression'].value_counts()
# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(depression_counts, labels=depression_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel', len(depression_counts)), startangle=90)
    
plt.title('Target Variable Distribution')
plt.show()
```

The 80%/20% class distribution is an important insight that suggests a significant class imbalance. This should be taken into consideration during preprocessing and model training to improve the model's ability to correctly predict the minority class. Techniques like resampling, changing evaluation metrics, and adjusting algorithm parameters can help make the model more effective at handling imbalanced data.

#### 1.1 Categorical features

```python
# Convert object columns to category 
df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).astype('category')
```

```python
for feature in df.columns:
     if df[feature].dtype == 'category':
        # Create a 1x2 subplot
        fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    
        # Count plot for the column
        sns.countplot(x=feature, data=df, ax=ax[0], hue=feature, palette='pastel', dodge=False)
        if ax[0].legend_:
            ax[0].legend_.remove()
        ax[0].set_title(f"Count plot of {feature}")
        ax[0].set_xlabel(feature)
        ax[0].set_ylabel("Count")
    
        # Add percentage labels on each bar
        total = len(df)
        for p in ax[0].patches:
            percentage = f'{(p.get_height() / total) * 100:.1f}%'
            ax[0].annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()), 
                           ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                           textcoords='offset points')
    
        # Prepare data for the stacked percentage plot
        if 'Depression' in df.columns:  # Ensure 'Depression' column exists
            crosstab = pd.crosstab(df[feature], df['Depression'], normalize='index') * 100
            bottom = pd.Series(0, index=crosstab.index)
    
            for category in crosstab.columns:
                ax[1].bar(crosstab.index, crosstab[category], bottom=bottom, label=f"{category}", color=sns.color_palette('pastel')[crosstab.columns.get_loc(category)])
                for idx, value in enumerate(crosstab[category]):
                    if value > 0:
                        ax[1].text(idx, bottom.iloc[idx] + value / 2, f"{value:.0f}%", ha='center', va='center', fontsize=10)
                bottom += crosstab[category]
    
            ax[1].set_title(f"Stacked Percentage of {feature} by Depression")
            ax[1].set_ylabel("Percentage")
            ax[1].legend(title='Depression', bbox_to_anchor=(1.05, 1), loc='upper left')
            ax[1].set_xticks(range(len(crosstab.index)))
            ax[1].set_xticklabels(crosstab.index)
    
        # Adjust layout for better spacing
        plt.tight_layout()
        plt.show()
```

#### 1.2 Numerical Features

```python
for feature in df.columns:
    if df[feature].dtype != 'category':
        plt.figure(figsize=(8,4))
        sns.histplot(data=df, x=feature, hue='Depression', kde=True, palette='viridis')
        plt.title(f'{feature} Distribution by Depression')
        plt.xlabel(feature)
        plt.ylabel('Count')
        plt.show()
```

```python
def age_group(df, age_column='Age'):
    # Define bins for the age ranges: 17-19, 20-22, ..., 57-60
    bins = [17, 20, 23, 26, 29, 32, 35, 38, 41, 44, 47, 50, 53, 56, 61]
    labels = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]

    # Generate labels as ranges in the format 'min-max'
    labels = [f'{bins[i]}-{bins[i+1]-1}' for i in range(len(bins) - 1)]  # Subtract 1 to avoid overlap

    # Create the new column 'Age_Group'
    df['Age_Group'] = pd.cut(df[age_column], bins=bins, labels=labels, right=False, include_lowest=True)

    # Convert 'Age_Group' to an ordered categorical type
    df['Age_Group'] = pd.Categorical(df['Age_Group'], categories=labels, ordered=True)
    return df
age_group(df)
fig, ax = plt.subplots(1, 2, figsize=(14, 4))

# Plot 1: Countplot for Age_Group
sns.countplot(data=df, x='Age_Group', ax=ax[0])
ax[0].set_title('Countplot of Age_Group')
ax[0].tick_params(axis='x', rotation=90)
ax[0].set_xlabel('Age Group')
ax[0].set_ylabel('Count')

# Prepare data for the stacked percentage plot
crosstab = pd.crosstab(df['Age_Group'], df['Depression'], normalize='index') * 100  # Normalize to percentages
bottom = pd.Series(0, index=crosstab.index)  # Initialize bottom tracker for stacking

# Plot 2: Stacked percentage bar plot
for category in crosstab.columns:
    ax[1].bar(crosstab.index, crosstab[category], bottom=bottom, label=f"{category}")
    # Add percentage annotations
    for idx, value in enumerate(crosstab[category]):
        if value > 0:  # Annotate only non-zero percentages
            ax[1].text(idx, bottom.iloc[idx] + value / 2, f"{value:.0f}%", ha='center', va='center')
    bottom += crosstab[category]  # Update the bottom for the next segment

# Set title, labels, and legend for the second plot
ax[1].set_title('Age_Group by Depression (Stacked %)')
ax[1].set_xlabel('Age Group')
ax[1].set_ylabel('Percentage')
ax[1].legend(title='Depression', bbox_to_anchor=(1.05, 1), loc='upper left')
ax[1].tick_params(axis='x', rotation=90)

# Adjust layout
plt.tight_layout()
plt.show()
```

The majority of participants are in the 17–19 age group, with a smaller proportion in the 21–24 group and very few in older age groups. Depression prevalence is higher among the 23–25 age group (29%) compared to the 17–19 group (21%), while older age groups report no cases of depression, likely due to underrepresentation.

# 2. Correlation

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

X = df.copy()
y = X.pop("Depression")

# Label encoding for categoricals
for colname in X.select_dtypes("category"):
    X[colname], _ = X[colname].factorize()

# All discrete features should now have integer dtypes (double-check this before using MI!)
discrete_features = X.dtypes == int

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_classif(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X, y, discrete_features)
mi_scores
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(10, 7))
plot_mi_scores(mi_scores)
```

### Key Observations from the Chart
Top Features:
Marital Status has the highest mutual information score, indicating it is the most influential feature in predicting the target variable.
Working Professional or Student and Study/Job Satisfaction are the next most significant features, suggesting their strong relevance.

Moderate Importance:
Features like Work/Study Hours, Social Media Usage Hours, Profession, and Suicidal Thoughts have moderate scores, showing their importance but less than the top features.

Lower Importance:
Features like Financial Stress, Loneliness Frequency, Self-Comparison, and Academic/Work Pressure show lower scores but still hold some predictive power.

Least Important Features:
Family Mental Illness History, Sleep Duration, Alcohol Consumption, and Smoking have minimal scores, indicating they contribute very little to the target variable prediction.

Health-Related Variables:
Surprisingly, variables like Physical Activity Frequency and Dietary Habits, often considered crucial in mental health studies, show lower significance here.

### Spearman Correlation Heatmap for Feature Relationships with Depression

```python
check = X.copy()
check['Depression'] = y

le = LabelEncoder()
check['Depression'] = le.fit_transform(check['Depression'])
corr_matrix = check.corr(method='spearman')

# Mask the lower triangle (set it to NaN)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask, fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap (Upper Triangle)")
plt.show()
```

The heatmap shows a strong correlation between "Profession" and "Working Professional or Student" (0.95), while other variables generally have weak to moderate associations, with most mental health variables showing negligible correlations with others.

### Cramér's V Analysis and Heatmap of Categorical Feature Associations

```python
import scipy.stats as ss

# Function to calculate Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, p, dof, expected = ss.chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))
category_cols = ['Gender','Sleep Duration','Working Professional or Student','Profession','Study Degree','Dietary Habits','Suicidal Thoughts','Family Mental Illness History','Depression']
# Create a matrix for Cramér's V between categorical columns
cramers_v_matrix = pd.DataFrame(index=category_cols, columns=category_cols)

for col1 in category_cols:
    for col2 in category_cols:
        cramers_v_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

# Display Cramér's V matrix
# Ensure the values in the DataFrame are numeric
cramers_v_matrix = cramers_v_matrix.astype(float)

# Plot the heatmap
mask = np.triu(np.ones_like(cramers_v_matrix, dtype=bool))
plt.figure(figsize=(6, 4))
sns.heatmap(cramers_v_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True,mask=mask)
plt.title("Cramér's V Heatmap")
plt.show()
```
